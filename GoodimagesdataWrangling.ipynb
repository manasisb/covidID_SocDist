```python
import glob as glob
import os 
import shutil
```

```python
jpgfiles = glob.glob("/Users/manasiweginwar/Desktop/Good-jpg/*.jpg")
```

```python
jpgfiles
```

```python
for file in jpgfiles:
    oldname = os.path.basename(file)
    oldpath = os.path.dirname(file)
#     print(oldname)
    if 'Good' not in oldname:
        newfilename = 'Good' + oldname
#         print(oldname)
#         print(newfilename)
        newfilenamePath = os.path.join(oldpath, newfilename)
        print(file, newfilenamePath)
        shutil.move(file, newfilenamePath)
```

```python
import xml.dom.minidom as md 
```

```python
xmlfiles = glob.glob("/Users/manasiweginwar/Desktop/Good-xml/*.xml")
```

```python
xmlfiles
```

```python
for filename in xmlfiles:
    file = md.parse(filename)
    oldvalue = file.getElementsByTagName("filename")[0].childNodes[0].nodeValue
    newval = "Good" + oldvalue
    file.getElementsByTagName("filename")[0].childNodes[0].nodeValue = newval
    print("filename:"+file.getElementsByTagName("filename")[0].childNodes[0].nodeValue)
    #edit path
    oldpath = file.getElementsByTagName("path")[0].childNodes[0].nodeValue
    newpath = os.path.join(os.path.dirname(oldpath), "Good"+oldpath.split('/')[-1])
    # newpath = "Caution" + oldpath
    file.getElementsByTagName("path")[0].childNodes[0].nodeValue = newpath
    print("pathname:"+ file.getElementsByTagName("path")[0].childNodes[0].nodeValue)
#     file.getElementsByTagName("path")[0].childNodes[0].nodeValue = newpath
#     with open(filename, "w" ) as fs:  
#         fs.write( file.toxml() ) 
#         fs.close() 
```
